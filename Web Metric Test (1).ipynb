```python
import requests

# Your API key from Google Cloud
API_KEY = '' #https://console.cloud.google.com/

# The URL you want to analyze
url = 'https://www.ubisoft.com/'

# API endpoint
pagespeed_api_url = 'https://www.googleapis.com/pagespeedonline/v5/runPagespeed'

# Parameters for the API request
params = {
    'url': url,
    'key': API_KEY,
    'strategy': 'mobile'  # You can also use 'mobile' for mobile metrics
}

# Make the API request
response = requests.get(pagespeed_api_url, params=params)
result = response.json()

# Extract relevant data
performance_score = result['lighthouseResult']['categories']['performance']['score'] * 100
first_contentful_paint = result['lighthouseResult']['audits']['first-contentful-paint']['displayValue']
speed_index = result['lighthouseResult']['audits']['speed-index']['displayValue']
time_to_interactive = result['lighthouseResult']['audits']['interactive']['displayValue']
bounce_rate = result['lighthouseResult']['audits']['metrics']['details']['items'][0].get('bounceRate', 'N/A')
session_duration = result['lighthouseResult']['audits']['metrics']['details']['items'][0].get('averageSessionDuration', 'N/A')
conversion_rate = result['lighthouseResult']['audits']['metrics']['details']['items'][0].get('conversionRate', 'N/A')

# Display the results
print(f'Performance Score: {performance_score}')
print(f'First Contentful Paint: {first_contentful_paint}')
print(f'Speed Index: {speed_index}')
print(f'Time to Interactive: {time_to_interactive}')
print(f'Bounce Rate: {bounce_rate}')
print(f'Session Duration: {session_duration}')
print(f'Conversion Rate: {conversion_rate}')
```
